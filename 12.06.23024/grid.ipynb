```python
import matplotlib.pyplot as plt

#data
x=[1,2,3,4,5]
y1=[1,4,9,16,25]
y2=[6,8,9,12,14]

#create the plot
plt.plot(x,y1,linestyle='-',color='green',linewidth=3,marker='o',markersize=8,markerfacecolor="black")
plt.plot(x,y2,linestyle='--',color='red',linewidth=3,marker='x',markersize=8,markerfacecolor="yellow")

#title and label
plt.title('customised line plot with grid')
plt.xlabel('x-axis')
plt.ylabel('y-axis')

# Add a legend
plt.legend()

#add the vcalue as grid
plt.grid(True)

#customise the grid
plt.grid(color='pink',linestyle='--',linewidth=1)

```
