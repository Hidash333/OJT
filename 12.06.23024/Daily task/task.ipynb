```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

file_path = 'weather_data(in).csv'
weather_data = pd.read_csv(file_path)

print(weather_data)
```

```python
# Step 1: Handle missing values
# Check for missing values
print(weather_data.isnull().sum())

# If there are missing values, you can drop them or impute them
# For example, to drop rows with missing values
weather_data.dropna(inplace=True)

# Step 2: Convert data types
# Convert "Date" column to datetime type
weather_data['Date'] = pd.to_datetime(weather_data['Date'])

# Step 3: Remove duplicates
# Remove duplicate rows
weather_data.drop_duplicates(inplace=True)

# Step 4: Check for outliers
# You can use descriptive statistics or visualization techniques to identify outliers

# Step 5: Feature engineering
# You can create new features based on existing ones if needed

# After performing these steps, you can check the cleaned and preprocessed DataFrame
print("Cleaned DataFrame:")
print(weather_data)
```

```python
# Calculate mean temperature
temperature_mean = np.mean(weather_data['Temperature'])

# Calculate standard deviation of temperature
temperature_std = np.std(weather_data['Temperature'])

# Calculate maximum temperature
temperature_max = np.max(weather_data['Temperature'])

# Calculate minimum temperature
temperature_min = np.min(weather_data['Temperature'])

# Print the calculated statistics
print("Mean temperature:", temperature_mean)
print("Standard deviation of temperature:", temperature_std)
print("Maximum temperature:", temperature_max)
print("Minimum temperature:", temperature_min)
```

```python
# Create a time series plot
plt.figure(figsize=(10, 6))  # Set the figure size

# Plot the temperature trend over time
plt.plot(weather_data['Date'], weather_data['Temperature'], color='blue', marker='o', linestyle='-')

# Add labels and title
plt.xlabel('Date')
plt.ylabel('Temperature (Celsius)')
plt.title('Temperature Trend Over Time')

# Rotate the x-axis labels for better readability
plt.xticks(rotation=45)

# Display grid
plt.grid(True)

# Show plot
plt.tight_layout()  # Adjust layout to prevent clipping of labels
plt.show()
```

```python
# Convert the 'Date' column to datetime and extract the month
weather_data['Date'] = pd.to_datetime(weather_data['Date'])
weather_data['Month'] = weather_data['Date'].dt.month

# Calculate the average monthly precipitation
average_monthly_precipitation = weather_data.groupby('Month')['Precipitation'].mean()

# Define months
months = ['January', 'February', 'March', 'April', 'May', 'June', 'July', 'August', 'September', 'October', 'November', 'December']

# Create a bar chart
plt.figure(figsize=(10, 6))
plt.bar(months, average_monthly_precipitation, color='skyblue')


# Add labels and title
plt.xlabel('Month')
plt.ylabel('Average Precipitation (mm)')
plt.title('Average Monthly Precipitation')

# Rotate x-axis labels for better readability
plt.xticks(rotation=45)

# Show plot
plt.tight_layout()
plt.show()
```

```python
# Create a scatter plot
plt.figure(figsize=(8, 6))
plt.scatter(weather_data['WindSpeed'], weather_data['Temperature'], color='blue', alpha=0.5)

# Add labels and title
plt.xlabel('Wind Speed (km/h)')
plt.ylabel('Temperature (Celsius)')
plt.title('Relationship Between Wind Speed and Temperature')

# Show plot
plt.grid(True)
plt.tight_layout()
plt.show()
```
