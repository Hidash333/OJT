```python
import matplotlib.pyplot as plt

#dataset
x = [1,2,3,4,5,6]
y = [2,3,4,5,7,11]

#create a plot with marker
plt.plot(x,y,marker='o',linestyle='-.',color='green',markersize=6,markerfacecolor='r')


#add the label as well as title
plt.title("Line plot with markers")
plt.xlabel("x-axis")
plt.ylabel("y-axis")

plt.show()
```
