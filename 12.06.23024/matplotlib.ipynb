```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
#dataset
x = [1,2,3,4,5,6]
y = [2,3,4,5,7,11]

#create a plot for data
plt.plot(x,y)

#customisation for the plot

#add a title
plt.title('Line Plot')

#Add a labels
plt.xlabel("x-axis")
plt.ylabel("y-axis")

#output the plot
plt.show()
```
