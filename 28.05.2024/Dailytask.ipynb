```python
import pandas as pd
df = pd.read_csv('air-pollution.csv')
print(df)
```

```python
#clean the data by removing rows that are empty
df_cleaned = df.dropna(how="all")

#fill the remaining nan values with 0
df_cleaned.fillna(0,inplace=True)

#remove duplicate entries
df_cleaned.drop_duplicates(inplace=True)

print(df_cleaned)
```

```python
#filtering based on country
filtered_data = df_cleaned[df_cleaned['Entity'] == 'India']
print(filtered_data)

```

```python
#to find median , mean and SD of a coloumn
filtered_data['median_Nitrogen oxide (NOx)'] = filtered_data['Nitrogen oxide (NOx)'].median()
filtered_data['mean_Nitrogen oxide (NOx)'] = filtered_data['Nitrogen oxide (NOx)'].mean()
filtered_data['std_Nitrogen oxide (NOx)'] = filtered_data['Nitrogen oxide (NOx)'].std()

filtered_data['median_Sulphur dioxide (SO₂) emissions'] = filtered_data['Sulphur dioxide (SO₂) emissions'].median()
filtered_data['mean_Sulphur dioxide (SO₂) emissions'] = filtered_data['Sulphur dioxide (SO₂) emissions'].mean()
filtered_data['std_Sulphur dioxide (SO₂) emissions'] = filtered_data['Sulphur dioxide (SO₂) emissions'].std()

filtered_data['median_Carbon monoxide (CO) emissions'] = filtered_data['Carbon monoxide (CO) emissions'].median()
filtered_data['mean_Carbon monoxide (CO) emissions'] = filtered_data['Carbon monoxide (CO) emissions'].mean()
filtered_data['std_Carbon monoxide (CO) emissions'] = filtered_data['Carbon monoxide (CO) emissions'].std()

print(filtered_data)
```
